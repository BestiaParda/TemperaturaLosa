# Temperatura en una losa de hormigón


## Objeto
El articulo trata de analizar la temperatura de una losa de hormigón expuesta a temperatura ambiente,

## Hipótesis de Partida
Se supondrá que la losa está expuesta a la temperatura exterior del aeropuerto Málaga.
El coeficiente de transmisión de calor corresponde a un flujo horizontal de aire de $4.5 m/s$ e incluirá la parte radiativa y convectiva, tomando como valor:
  $$h=25\frac{W}{m^2K}$$  

```python
file='Malaga.csv'
reader=csv.reader(open(file))        
L=[]  
x_hora=[]
y_temperatura=[]

for row in reader:
    L.append(row)
    
for row in L:
    x_hora.append(row[1])  
    y_temperatura.append(row[6]) 
    
 
graf1=line(zip(x_hora,y_temperatura),axes_labels =['$h$','$\degree$$C$'],legend_label='$T_{Malaga}$')
graf1.set_axes_range(0, 8760, 0, 40)
show (graf1)
```
